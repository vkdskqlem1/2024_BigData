```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt

# 한글폰트 적용하기
plt.rc('font',family = 'Malgun Gothic')

# 랜덤 라이브러리 불어괴
import random

# 데이터 만들기
height = []
weight = []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))

# 제목 : 키와 몸무게의 상관관계
plt.title('키와 몸무게의 상관관계')

# 산점도 그래프 그리기
plt.scatter(height,weight, c=height, cmap='RdPu')

# 축 제목
plt.colorbar(label='키')
plt.xlabel('키')
plt.ylabel('몸무게')

plt.show()
```

```python
import matplotlib.pyplot as plt

# 한글폰트 적용
plt.rc('font',family = 'Malgun Gothic')

# 제목
plt.title('오디션 프로그램 득표 현황')

# 값
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

for i in range(len(week1)):
    week1[i] =- week1[i]
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 표시    

# 출력 값
plt.barh(singer, week1, label = '첫째주', color = 'hotpink')
plt.barh(singer,week2,label='둘째주',color='royalblue')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 한글폰트 적용
plt.rc('font',family = 'Malgun Gothic')

# 제목
plt.title('오디션 프로그램 적용하기')

# 값
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

# 출력 값
plt.bar(range(1,6), week1, label = '첫째주', color = 'hotpink')
plt.bar(range(7,12), week2, label = '둘째주', color = 'royalblue')
plt.xticks(range(1,12),['A','B','C','D','E','','A','B','C','D','E'])
plt.legend()
plt.show
```

```python
import matplotlib.pyplot as plt

# 데이터
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

# 한글폰트 적용하기
plt.rc('font',family = 'Malgun Gothic')

# 제목
plt.title('오디션 프로그램 득표 현황')

# 출력 값
plt.bar(range(1,14,3), week1,label = '첫째 주' ,color = 'hotpink')
plt.bar(range(2,15,3), week2, label = '둘째 주' , color = 'royalblue')
plt.xticks(range(1,14,3),singer)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

import random

h1,w1,s1,h2,w2,s2 = [],[],[],[],[],[]

for i in range(100):
    h1.append(random.randint(140,180))
    w1.append(random.randint(40,80))
    s1.append(random.randint(0,200))
    h2.append(random.randint(160,200))
    w2.append(random.randint(50,100))
    s2.append(random.randint(0,200))
    
# 한글폰트 적용하기
plt.rc('font',family = 'Malgun Gothic')

# 제목
plt.title('키와 몸무게, 체육 점수의 상관관계')

plt.scatter(h1,w1,s=s1, color='crimson', alpha=0.7, label='그룹1')
plt.scatter(h2,w2,s=s2, color='indigo', alpha=0.7, label = '그룹2')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.legend()
plt.show()
```
