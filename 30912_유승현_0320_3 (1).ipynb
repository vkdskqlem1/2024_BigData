```python
print('a부터 b까지 정수의 합을 구합니다.')
a = int(input('정수 a를 입력하세요. :'))
b = int(input('정수 b를 입력하세요. :'))

if a > b:
    a, b = b, a

sum = 0 
for i in range(a, b+1):
    if i < b:
        print(f'{i} +',end="")
    else:
        print(f'{i} = ',end="")
    sum = sum + i

print(sum)
```

```python
print('a부터 b까지 정수의 합을 구합니다.')
a = int(input('정수 a를 입력하세요. :'))
b = int(input('정수 b를 입력하세요. :'))

if a > b:
    a, b = b, a

sum = 0 

for i in range(a, b):
    print(f'{i} +',end = "")
    sum = sum + i

print(f'{b} = ', end="")
sum = sum + b

print(sum)
```

```python
print('+와 -를 번갈아 출력합니다')

n = int(input('몇 개를 출력할까요?: '))

for i in range(n):
    if i % 2:
        print('-',end="")
    else:
        print('+',end="")
print()
```
