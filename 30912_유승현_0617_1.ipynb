```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt

# 한글폰트 적용하기
plt.rc('font',family = 'Malgun Gothic')

# data
sales = [1,5,7,3,7]
month = [3,4,5,6,7]

# 제목
plt.title('월별 판매 실적')

# 가로 비차트 그리기 barh range(시작, 마지막+1)
plt.barh(range(3,8),sales,color='red')
plt.show()
```

## 원형차트 그리기 

```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt

# 한글폰트 적용하기
plt.rc('font',family='Malgun Gothic')

# data 혈액형 수
b_type = [25,19,37,11]
b_name = ['A형','B형','C형','D형']
# 제목 : 혈액형 분포
plt.title('혈액형 분포')
# 원형차트 그리는 명령어 pie 범례 labels = [] autopct = ''
plt.pie(b_type, labels = b_name, autopct = '%1.1f%%')
plt.show()
```

```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt

# 한글폰트 적용하기
plt.rc('font',family='Malgun Gothic')

# data 국어 영어
kor = [80,20,50,20,10,50,60,30,40]
eng = [90,40,10,40,60,20,50,70,90]

# 산점도 그리기 scatter option color = 'r' 투명도 alpha 값
plt.scatter(kor,eng,color='r',alpha=0.3)
# 축 제목 x,y축
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
```
