```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font',family = 'Malgun Gothic')

# data
sales = [1,5,7,3,7]
month = [3,4,5,6,7]

# 막대 그래프 그리기 plt.bar()
plt.title('월별 판매 실적') # 제목
plt.bar(month,sales,color='b') # 컬러 파란색

plt.show()
```

```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font',family = 'Malgun Gothic')

# data
sales = [1,5,7,3,7]
month = [3,4,5,6,7]

# 제목
plt.title('월별 판매 실적')

# 가로 빅차트 그리기 barh(시작, 마지막+1)
plt.barh(range(3,8),sales,color = 'r')
plt.show()
```
